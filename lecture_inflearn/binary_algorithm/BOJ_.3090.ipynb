```python
# 들어가기 앞서서 차이의 최댓값을 X 이하로 만들기위해서 연산이 적어도 몇개가 필요한지 먼저 해보자

# example
X = 3
a = [10, 6, 5, 1, 9, 7, 2, 3, 10, 5, 3]

# 1. 왼쪽에서 오른쪽으로 
# 만약 (왼쪽 << 오른쪽)이라면 줄여나가는 식으로 진행

# a = [10, 6, 5, 1, 9 ..] 9가 나온 시점에 왼쪽보다 오른쪽이 더 큼 -> 오른쪽에 있는것을 줄여준다 (즉, 9 -> 4) 연산 5번 사용함

# 다시 a = [10, 6, 5, 1, 4, 7, 2, 3, 10] 10이 나온 시점에 마찬가지로 더 크기 때문에 줄여준다. (즉 10 -> 6) 연산 4번 사용

# 더이상 만족하는게 없음 a = [10, 6, 5, 1, 4, 7, 2, 3, 6, 5, 3]

# 2. 이제 (왼쪽 >> 오른쪽)경우를 생각해보자. 오른쪽에서 왼쪽으로 읽자

# a = [3, 5, 6, 3, 2, 7->5, 4, 1, 5->4, 6, 10->9] 총 4번 사용

# 결론은 13번이 필요하다라고 답을 얻었음.

# 구현
count = 0
test_a = [10, 6, 5, 1, 9, 7, 2, 3, 10, 5, 3]
x = 3

# forward
for i in range(1,len(test_a)):
    if (test_a[i] - test_a[i-1]) > x:
        temp = test_a[i] - x - test_a[i-1]
        test_a[i] = test_a[i] - temp
        count = count + temp
print(test_a)

# reverse
for i in range(len(test_a)-1, 0, -1):
    if (test_a[i-1] - test_a[i]) > x:
        temp = test_a[i-1] -x - test_a[i]
        #print(temp)
        test_a[i-1] = test_a[i-1] - temp
        count = count + temp
print(test_a)
```

```python
# 주어진 건 수정할수있는 수가 주어짐.

def cal(x: int, limit_count: int, a: list) -> list and int and bool:
    meet = False
    ans = []
    count = 0
    for i in range(1, len(a)):
        if (a[i] - a[i-1]) > x:
            temp = a[i] - x - a[i-1]
            a[i] = a[i] - temp
            count = count + temp
            
    for i in range(len(a)-1, 0, -1):
        if (a[i-1] - a[i]) > x:
            temp = a[i-1] - x - a[i]
            a[i-1] = a[i-1] - temp
            count = count + temp
    ans = a
    if limit_count >= count:
        # 아직 count가 남았을때
        meet = True
    else:
        # 이미 오버가 되었을때
        meet = False
    
    return ans, (limit_count - count), meet

N, T = map(int, input().split())
test_b = list(map(int, input().split()))

#for i in range(N):
#    test_b.append(int(input()))

limit_count = T
#diff_list = []
#for i in range(1, len(test_b)):
#    diff_list.append(abs(test_b[i-1] - test_b[i]))
low = 0
high = int(1e9) #max(diff_list)

answer = []
temp_count = limit_count
temp_ans = test_b
backup = temp_ans
while low <= high:
    mid = int((low+high)/2)
    backup = temp_ans
    temp_ans, left_count, temp_meet = cal(mid, temp_count, temp_ans)

    if temp_meet == True and left_count == 0: # 오버 하지 않았을때 (아직 잔여 count가 남아서 더 연산가능)
        answer = temp_ans
        break
    
    elif temp_meet == True and left_count != 0:
        temp_count = left_count
        high = mid
        temp_ans = temp_ans
        
    else: # 오버했을때 False
        low = mid + 1
        temp_count = limit_count
        temp_ans = backup

print(answer)
```

```python
N, T = tuple(map(int, input().split(" ")))
A = list(map(int, input().split(" ")))

def needed_num_operation(x: int) -> int:
    B = [A[i] for i in range(N)]
    
    num_operation = 0
    for i in range(N-1):
        if B[i] + x < B[i+1]:
            num_operation += B[i + 1] - B[i] - x
            B[i + 1] = B[i] + x
    
    for i in range(N-1, 0, -1):
        if B[i] + x < B[i - 1]:
            num_operation += B[i-1] - B[i] - x
            B[i-1] = B[i] + x
    
    return num_operation

low = 0
high = int(1e9)
answer = -1      

while low <= high:
    mid = int((low + high)/2)
    
    if needed_num_operation(mid) <= T:
        answer = mid
        high = mid - 1
    else:
        low = mid + 1
        
for i in range(N-1):
    if A[i] + answer < A[i+1]:
        A[i + 1] = A[i] + answer
        
for i in range(N-1, 0, -1):
    if A[i] + answer < A[i-1]:
        A[i-1] = A[i] + answer
        
print(" ".join(list(map(str, A)))) 
```
