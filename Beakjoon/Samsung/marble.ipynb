```python
from collections import deque
n, m = map(int, input().split())
maze = [list(map(int, input().split())) for _ in range(n)]

for i in range(n):
    for j in range(m):
        if maze[i][j] == 'R':
            rx, ry = i,j
        if maze[i][j] == 'B':
            bx, by = i, j

dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

def bfs(rx, ry, bx, by):
    global maze
    Q = deque()
    Q.append((rx, ry, bx, by))
    
    # 방문 여부를 판단
    visited = []
    visited.append((rx, ry, bx, by))
    count = 0
    
    while Q:
        for _ in range(len(Q)):
            rx, ry, bx, by = Q.popleft()
            if count > 10:
                print(-1)
                return
            
            if maze[rx][ry] == 'O':
                print(count)
                return
            
            for i in range(4):
                nrx = rx
                nry = ry
                
                while True:
                    nrx += dx[i]
                    nry += dy[i]
                    
                    if maze[nrx][nry] == '#':
                        nrx -= dx[i]
                        nry -= dy[i]
                        break
                    
                    if maze[nrx][nry] == 'O':
                        break
                
                nbx = bx
                nby = by
                
                while True:
                    nbx += dx[i]
                    nby += dy[i]
                
                    if maze[nbx][nby] == '#':
                        nbx -= dx[i]
                        nby -= dy[i]
                        break
                    
                    if maze[nbx][nby] == 'O':
                        break
                
                if maze[nbx][nby] == 'O':
                    continue
                
                if nrx == nbx and nry == nby: # 두 구슬의 위치가 같을때
                    if abs(nrx - rx) + abs(nry - ry) > abs(nbx - bx) + abs(nby - by): 
                        nrx -= dx[i]
                        nry -= dy[i]
                    else:
                        nbx -= dx[i]
                        nby -= dy[i]
                if (nrx, nry, nbx, nby) not in visited:
                    Q.append((nrx, nry, nbx, nby))
                    visited.append((nrx, nry, nbx, nby))
                
        count += 1
    print(-1)
    
bfs(rx, ry, bx, by)
                
            

```
