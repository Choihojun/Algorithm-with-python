```python
from collections import deque
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]
dice = [6,4,2,5,3,1] # down left front back right up
# dlfbru
def move_right():
    global dice
    temp = dice[1]
    dice[1] = dice[0]
    dice[0] = dice[4]
    dice[4] = dice[5]
    dice[5] = temp

def move_left():
    global dice 
    temp = dice[4]
    dice[4] = dice[0]
    dice[0] = dice[1]
    dice[1] = dice[5]
    dice[5] = temp

def move_front():
    global dice
    temp = dice[3]
    dice[3] = dice[0]
    dice[0] = dice[2]
    dice[2] = dice[5]
    dice[5] = temp

def move_back():
    global dice
    temp = dice[5]
    dice[5] = dice[2]
    dice[2] = dice[0]
    dice[0] = dice[3]
    dice[3] = temp


def dice_pos_value(td):
    # 목표 방향으로 이동후 위치 갱신 및 주사위 값 계산 
    global dice_pos, target_direction, current_dice_value, dice
    print("td", td)
    # 위치 갱신
    xx = dice_pos[0] + dy[td]
    yy = dice_pos[1] + dx[td]
    
    #if board[yy][xx]:
    if 0 <= xx < N and 0 <= yy < M:
        dice_pos[0] = xx
        dice_pos[1] = yy
    
    else: # 막혔을경우 반대로 간다
        td = td - 2
        #if td < 0:
        #    td = td + 4
        dice_pos[0] = dice_pos[0] + dy[td]
        dice_pos[1] = dice_pos[1] + dx[td]
    
    print(dice_pos)
    
    # 값 계산 (TODO) -> 주사위를 만들어야함!
    if td == 0:
        move_left()
    elif td == 1:
        move_front()
    elif td == 2:
        move_right()
    elif td == 3:
        move_back()
        
    #dlfbru
    current_dice_value = dice[0]
    
    return current_dice_value
    
    
def dice_direction(d):
    global dice_pos
    A = dice_pos_value(d)
    B = board[dice_pos[0]][dice_pos[1]]
    print("A: ", A, "B: ", B)
    if A > B:
        d = d + 1 # 90도 시계방향 회전 (d = dx 또는 dy의 인덱스)
        if d == 4:
            d = d % 4
    if A < B:
        d = d - 1 # 90도 반시계방향 회전
        if d == -1: # d가 -1인 경우
            d = 3
    if A == B:
        d = d
    return d

def calcul_score(): # Okay!
    # BFS로 구현을 해야함
    global dice_pos, board
    visited = [[0] * M for _ in range(N)]
    #print("score dice pos", dice_pos)
    visited[dice_pos[0]][dice_pos[1]] = 1
    
    return_value = board[dice_pos[0]][dice_pos[1]]
    Q = deque()
    Q.append((dice_pos[0], dice_pos[1]))
    
    while Q:
        tmp_pos = Q.popleft()
        #print("tmp_pos!", tmp_pos)
        cur_pos_value = board[tmp_pos[0]][tmp_pos[1]]
        for i in range(4):
            xx = tmp_pos[0] + dy[i]
            yy = tmp_pos[1] + dx[i]
            if 0 <= xx < N and 0 <= yy < M and (cur_pos_value == board[xx][yy]) and visited[xx][yy] == 0:
                return_value += board[xx][yy]
                visited[xx][yy] = 1
                Q.append((xx,yy))
    
    #print("return_value", return_value)
    return return_value
        
# 입력 받기
N, M, k = map(int, input().split()) # 세로 가로 이동 횟수

board = list()

board = [[4, 1, 2, 3, 3], 
         [6, 1, 1, 3, 3], 
         [5, 6, 1, 3, 2], 
         [5, 5, 6, 5, 5]]
#for _ in range(N):
#    tmp = list(map(int, input().split()))
#    board.append(tmp)
    
score = 0
current_dice_value = 3
target_direction = 2
dice_pos = [0, 0]

for _ in range(k):
    
    target_direction = dice_direction(target_direction)
    score = score + calcul_score() 
    
    
print(score)
```
