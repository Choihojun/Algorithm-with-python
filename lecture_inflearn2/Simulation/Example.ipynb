```python
import turtle as t
import random as r

dy = [-1, 0, 1, 0] #여기서는 y가 행이라서 그럼
dx = [0, 1, 0, -1]

class Brick():
    def __init__(self): # 생성자
        self.y = 0
        self.x = 6
        self.color = r.randint(1, 6)
    def move_left(self, grid):
        if grid[self.y][self.x-1] == 0 and grid[self.y+1][self.x - 1] == 0: # 빈공간일때
            grid[self.y][self.x] = 0
            self.x = self.x -1 
    
    def move_right(self, grid):
        if grid[self.y][self.x + 1] == 0 and grid[self.y+1][self.x + 1] == 0:
            grid[self.y][self.x] = 0
            self.x = self.x + 1
            

def draw_grid(block, grid):
    block.clear() #이것도 마찬가지임
    # 시작점
    top = 250
    left = -150
    colors = ['black', 'red','blue', 'orange','yellow', 'green', 'purple','white']
    for y in range(len(grid)): # 행 개수가 몇개냐를 의미함 0~24
        for x in range(len(grid[0])):
            sc_x = left + (x * 22)
            sc_y = top - (y * 22)
            block.goto(sc_x, sc_y)
            block.color(colors[grid[y][x]])
            block.stamp()
            
def DFS(y, x, grid, color):
    global ch, blank
    ch[y][x] = 1 #현재 위치
    blank.append((y,x))
    for i in range(4):
        yy = y + dy[i]
        xx = x + dx[i]
        if 0 < yy < 24 and 0 < xx < 13: # 경계선
            if grid[yy][xx] == color and ch[yy][xx] == 0:
                DFS(yy, xx, grid, color)

def grid_update(grid, blank):
    # blank에 있는 좌표들은 제거해야함
    for y, x in blank:
        grid[y][x] = 0
    for y in range(23, 0, -1):
        for x in range(1, 13):
            if grid[y][x] == 0: # 빈공간이 발견됬을때
                tmp_y = y
                while grid[tmp_y-1][x] == 0 and tmp_y - 1 > 0:
                    tmp_y -= 1
                grid[y][x] = grid[tmp_y-1][x]
                grid[tmp_y-1][x] = 0                 
    
    # 지우고 나서 떨어뜨리기 -> 전략 (가로를 쭉 검색하다가 0이 나왔을때 (빈공간), 이때 위로 탐색을 수행한 후 위에서 하나씩 떨어뜨린다.)


if __name__ == "__main__":
    sc = t.Screen()
    sc.tracer(False) # 더 빠르게 그림
    sc.bgcolor("black")
    sc.setup(width=600, height=700)
    
    # 격자판
    grid = [[0] * 12 for _ in range(24)]
    
    # 벽생성하기
    for i in range(24):
        grid[i].insert(0, 7)
        grid[i].append(7)
    grid.append([7] * 14)
    for y in range(23, 20, -1):
        for x in range(1, 13):
            grid[y][x] = r.randint(1, 6)
    
    # 움직이는 객체
    block = t.Turtle()
    block.penup() # 이걸 해야 움직일때 그림을 안그림 (궤적을 안그림)
    block.speed(0) # 속도는 0이 제일 빠름
    block.color("red")
    block.shape('square')
    # 여기서 중요한점은 buffer를 제거해줘야 동일한 속도로 떨어지게됨
    block.setundobuffer(None)
    
    
    #draw_grid(block, grid)
    brick = Brick() # 객체 생성
    grid[brick.y][brick.x] = brick.color
    import time
    
    
    # 키보드 입력
    sc.onkeypress(lambda: brick.move_left(grid), "Left")
    sc.onkeypress(lambda: brick.move_right(grid), "Right")
    sc.listen()
    
    
    while True:
        sc.update() #이걸 해줘야 while문 돌면서 그림이 갱신됨
        
        # 멈추는 코드도 필요함
        if grid[brick.y+1][brick.x] == 0:
            grid[brick.y][brick.x] = 0
            brick.y += 1
            grid[brick.y][brick.x] = brick.color
        else:
            # 인접한 같은색 블록 개수 파악하기
            ch = [[0] * 14 for _ in range(25)]
            blank = []
            DFS(brick.y, brick.x, grid, brick.color)
            
            # 블록 개수에 따른 제거 및 중력으로 떨어뜨리는 부분
            if len(blank) >= 4:
                grid_update(grid, blank)
            print(len(blank))
            
            brick=Brick()
        #for x in grid:
        #    print(x)
        #print()
        draw_grid(block, grid)
        time.sleep(0.02)
    
    draw_grid(block, grid)
    for x in grid:
        print(x)
    draw_grid(block, grid)    
    sc.mainloop()
```
